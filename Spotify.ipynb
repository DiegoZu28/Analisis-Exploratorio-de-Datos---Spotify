Pregunta 3.

Cuales son los tipos de album mas comunes en Spotify?

Cuantas canciones pertenecen a ese tipo?

```python
tipo_album = spotify['Album_type'].value_counts().reset_index()
tipo_album
```

```python
plt.figure(figsize = (10, 8))
plt.pie(tipo_album['count'], labels = tipo_album['Album_type'], autopct = '%1.1f%%', shadow = True, explode = (0.1, 0, 0), startangle = 75, colors = 'brg')
plt.title('Tipos de albumes')
plt.show()
```

Pregunta 4.

Como el Promedio de vistas, likes y comentarios se comparan entre los distintos tipos de albumes?

```python
df = spotify.groupby('Album_type')[['Views', 'Likes', 'Comments']].mean().reset_index()
df
```

```python
df_melt = pd.melt(df, id_vars = 'Album_type', var_name = 'Atributo', value_name = 'Total')
df_melt
```

```python
plt.figure(figsize = (10, 8))
sns.barplot(x = 'Album_type', y = 'Total', hue = 'Atributo', data = df_melt)
plt.title('Comparacion entre los tiopos de albumes')
plt.show()
```

Pregunta 5.

Top 5 canales de Youtube basado en vistas
